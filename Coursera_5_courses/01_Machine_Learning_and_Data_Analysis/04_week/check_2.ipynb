```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Воспользуемся документацией на модуль stats (https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html), выберем наугад какое либо распределение для непрерывной случайной величины.
Наугад выбралось: Распределение Gompertz (судя по имени - немец.....как тут не вспомнить "Формула любви")
Описание распределения - https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.gompertz.html#scipy.stats.gompertz
Что оно из себя представляет? Не вопрос, построим график плотности вероятности (Probability Density Function - pdf), для этого:
1. Генерируем массив значений случайной величины
2. Производим вычисление значения функции pdf для сформированных значений
3. Собственно строим график.



```python
x = np.linspace(0,10,1000)
c = 0.01
gmp = sts.gompertz(c)
gmp_pdf = gmp.pdf(x)
gmp_mean, gmp_var = sts.gompertz.stats(c, moments='mv')

nrm = sts.norm(gmp_mean, gmp_var)
nrm_pdf = nrm.pdf(x)

plt.plot(x, gmp_pdf)
plt.plot(x, nrm_pdf)
plt.ylabel('$P(X=x)$')
plt.xlabel('$x$')
plt.title('Density of a probability for the Gompertz and Norm distributions',fontsize=11)
```

Оценим выборочное среднее для данных 1000 выборок. Выборочное среднее высчитывается как среднее арифметическое.

Но для начала, определим вспомогательные функции, создающие, собственно, набор и определяющие среднее значение.

Используя информацию о среднем и дисперсии исходного распределения (её можно без труда найти в википедии), посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. Поверх каждой гистограммы нарисуйте плотность соответствующего нормального распределения (будьте внимательны с параметрами функции, она принимает на вход не дисперсию, а стандартное отклонение).

Опишите разницу между полученными распределениями при различных значениях n. Как меняется точность аппроксимации распределения выборочных средних нормальным с ростом n?

```python
def CreateSet(ElementsInSet, NumberOfSet):
    ArrayToReturn = np.zeros((NumberOfSet, ElementsInSet))
    set_rnd = gmp.rvs(size = ElementsInSet)
    for i in range(NumberOfSet):
        set_rnd = gmp.rvs(size = ElementsInSet)
        for j in range(ElementsInSet):
            ArrayToReturn[i][j] = set_rnd[j]
    return ArrayToReturn

def CalculateAverages(Set):
    rows = Set.shape[0]
    cols = Set.shape[1]
    Avrgs = np.zeros((rows, 1))
    for i in range(rows):
        summ = 0
        for j in range(cols):
            summ = summ + Set[i][j]
        Avrgs[i] = summ / cols
    return Avrgs
```

```python
Set_n5 = CreateSet(5, 1000)
Set_n10 = CreateSet(10, 1000)
Set_n50 = CreateSet(50, 1000)

Set_n5_avrg = CalculateAverages(Set_n5)
Set_n10_avrg = CalculateAverages(Set_n10)
Set_n50_avrg = CalculateAverages(Set_n50)
```

```python
x = np.linspace(1,7,1000)
nrm = sts.norm(gmp_mean, math.sqrt(gmp_var/5))
nrm_pdf = nrm.pdf(x)

plt.hist(Set_n5_avrg, bins = 20, normed = True)
plt.plot(x, nrm_pdf)
plt.ylabel('Number')
plt.xlabel('Value of the average in the sets')
plt.title('Histogram for the set with n = 5 elements VS Norm distrib',fontsize=14)
```
