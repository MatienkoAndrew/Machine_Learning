# 8.1. Композиция деревьев

Решающие деревья слишком легко подгоняются под обучающую выборку и получаются непригодными(переобученными) для построения прогнозов.

Однако решающие деревья очень хорошо подходят для объединения в композиции и построении одного не переобученного алгоритма на основе большого количества решающих деревьев/алгоритмов.

**Решающее дерево обладает следующими недостатками:**

- легко переобучается

- сильно меняется при небольшом значении выборки

Однако второй пункт можно превратить в достоинство благодаря композиции.

## 8.1.1. Композиция алгоритмов

**Композиция - это объединение $N$ алгоритмов $b_1(x), ..., b_N(x)$ в один**. Идея заключается в том, чтобы обучить алгоритмы, а затем усреднить полученные на них ответы:

$$\Large a(x) = \frac{1}{N}\sum_{n=1}^Nb_n(x)$$

это выражение будет ответом в задаче регрессии.

В задаче классификации нужно взять знак от выражения:


$$\Large a(x) = sign\frac{1}{N}\sum_{n=1}^Nb_n(x)$$

---
$\Large a(x) - композиция\ N\ алгоритмов\ b_1(x),...,b_N(x)$

$\Large b_1(x),...,b_N(x) - базовые\ алгоритмы$

## 8.1.3. Рандомизация

Чтобы построить композицию, нужно сначала обучить $N$ базовых алгоритмов, причем их нельзя обучать на
всей обучающей выборке, так как в этом случае они получаются одинаковыми, и в их усреднении не будет
никакого смысла.

Использовать ***рандомизацию***, то есть обучать базовые алгоритмы на разных подвыборках обучающей
выборки, — это один из способов сделать базовые алгоритмы различными. А поскольку решающие деревья сильно меняются даже от небольших изменений обучающей выборки, такая рандомизация значительно
повышает различность базовых алгоритмов.

1) Так называемый ***бутстрап*** — один из популярных подходов к построению подвыборок. Он заключается
в том, что из обучающей выборки длины $\ell$ выбирают с возвращением $\ell$ объектов. При этом новая выборка
также будет иметь размер $\ell$, но некоторые объекты в ней будут повторятся, а некоторые объекты из исходной выборки в нее не попадут. Можно показать, что в бутстрапированной выборке будет содержаться в
среднем 63% различных объектов исходной выборки.

2) Другой подход к рандомизации — ***генерация случайного подмножества обучающей выборки***. Размер этого
случайного подмножества является гиперпараметром. Например, можно случайно взять половину исходной
выборки и обучить на ней базовый алгоритм. Этот подход несколько проигрывает бутстрапу, так как содержит гиперпараметр, в то время как бутстрап без какой-либо настройки выдает подвыборку.
