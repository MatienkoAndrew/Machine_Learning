# Знакомство с машинным обучением

**Машинное обучение - извлечение закономерностей из примеров**

$X = (x_i,y_i)^l_{i=1}$ - обучающая выборка, где $x_i$ - $i$-ый объект, $y_i$-$i$-ый ответ

## Алгоритм

$a(x)$ - алгоритм, модель - это функция, которая переводит объекты в ответы

## Функционал ошибки

$Q(a, X)$ - ошибка алгоритма $a$ на выборке $X$

## Обучение

$Q(a, X) -> min$ - **задача обучения состоит в подборе такого алгоритма $a$, на котором достигается минимум функционала ошибки**

$[]$ - скобки Айверсона, если внутри них $True$, то возвращается $1$, иначе - $0$

## Основные вопросы машинного обучения

1) Каким выбрать функционал ошибки?

2) Какое взять семейство алгоритмов?

3) Как обучить алгоритм?

# Обучение без учителя

**Обучение с учителем** - это задачи, в которых у нас есть объекты и ответы на них. И нужно по этим парам восстановить нужную зависимость, построить алгоритм или модель, которые будут предсказывать ответы по объектам.

**Обучение без учителя** - это задача, в которой есть только объекты. И при этом с этими объектами нужно что-то сделать.

**Частичное обучение**  - есть объекты, и ответы лишь на часть объектов. Имея эту информацию, нужно восстановить общую зависимость.

**Активное обучение** - это задачи, в которых есть объекты, но в которых получать ответы очень дорого или тяжело. Поэтому алгоритм должен уметь определять на каких объектах ему надо знать ответ, чтобы лучше всего обучиться.

## Кластеризация

Есть объекты и нужно найти группы похожих объектов

Примеры:

1) Сегментация пользователей

2) Поиск схожих пользователей в социальных сетях

## Визуализация

Задача: нарисовать многомерную выборку

Должна отражать структуру объектов

Должно быть красиво

## Поиск аномалий

Задача: определить объект, который не похож на все остальные

# Признаки в машинном обучении

## Признаки

это то, что описывает объект

## Бинарные признаки

$D_j = {0, 1}$

Доход клиента выше среднего

## Вещественные признаки

$D_j = R$ - множество значений это все вещественные числа

## Категориальные признаки

$D_j$ - неупорядоченное множество

Можем сравнивать лишь на равенство, на совпадение. НО нельзя сравнивать на больше или меньше

Примеры:

1) Цвет глаз

2) Город

3) Образование (может быть упорядоченным)

Очень трудны в обращении

## Порядковые признаки

Частный случай категориального признака

$D_j$ - упорядоченное множество

Примеры:

1) Роль в фильме

2) Тип населенного пункта

## Множественные признаки

$D_j $ - множество всех подмножеств

Примеры:

1) Какие фильмы посмотрел пользователь?

2) Какие слова входят в текст?

# Линейные модели в задачах регрессии

## Линейная модель

$\Large a(x) = w_0 + \sum^d_{j=1}w_jx^j;$  

$\Large w_0$ - сдвиг, свободный коэффициент, член смещения

$\Large a(x) = \sum^{d+1}_{j=1}w_jx^j = <w, x>$ - к $w_0$ добавили $x_0 = 1$

## Отклонение признака

1) $a(x) - y$ - не подходит, потому что $min$ стремится к $-бесконечности$

2) $|a(x)-y|$ - модуль отклонения. $min$ достигается в $0$. Но модуль - это функция не гладкая. Градиента в нуле не существует.

3) $(a(x)-y)^2$ - квадрат отклонения. $min$ достигается в $0$. Функция гладкая, частная производная существует на всей функции. Можно минимизировать.

## Среднеквадратичная ошибка

$$\Large Q(a, X) = \frac{1}{l}\sum^l_{i=1}(a(x_i)-y)^2$$

Вычисляется квадрат отклонения ответа алгоритма от истинного ответа и усредняем по всей обучающей выборке.

$$\Large Q(a, X) = \frac{1}{l}\sum^l_{i=1}(<w,x_i>-y)^2$$
