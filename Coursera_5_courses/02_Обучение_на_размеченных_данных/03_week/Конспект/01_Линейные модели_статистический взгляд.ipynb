## Метод максимального правдоподобия

***Хорошее объяснение:***
https://www.youtube.com/watch?v=2iRIqkm1mug

Метод максимального правдоподобия говорит, что выбирайте ту гипотезу, при которой вероятность имеющихся наблюдений максимальна

ML - maximul likelihood

- Есть неизвестный параметр $\theta$

- Хотим построить оценку $\hat{\theta}$

- В качестве оценки неизвестного параметра $\theta$ возьмем такое значение $\hat{\theta}$ при котором вероятность имеющихся данных максимальна

Пример:

$y_1=0, y_2=1,y_3=2, y_4=0$

$y_i$ - независимы



$\hat{p}_{ML}-?$

$P(y_1=0,y_2=1,y_3=2,y_4=0)$

Так как $y$ - независимы:

$P(y_1=0)P(y_2=1)P(y_3=2)P(y_4=0) = p\cdot 2p\cdot (1-3p)\cdot p = 2p^3(1-3p)$ - нужно максимизировать, подбирая параметр $p$

***Логарифмируем функцию, так как максимизировать сумму легче, чем произведение***

$l(p) = ln(2p^3(1-3p)) = ln2 + 3lnp + ln(1-3p)$

$l'(p) = 3\cdot \frac{1}{p} - 3\cdot \frac{1}{1-3p}$

$l'(p) = 0$

$3\cdot \frac{1}{\hat{p}} - 3\cdot \frac{1}{1-3\hat{p}}=0$

$1-3\hat{p} - \hat{p} = 0$

$4\hat{p}=1$

$\hat{p} = \frac{1}{4}$ - получили оценку параметра $p$ методом максимального правдоподобия
