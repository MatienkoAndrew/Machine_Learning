```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
prices_data = pd.read_csv('close_prices.csv')
prices_data
```

```python
X = prices_data.iloc[:, 1:]
X
```

```python
clf = PCA(n_components=10)
clf.fit(X)
```

```python
sort_disp = sorted(clf.explained_variance_ratio_, reverse=True)
sort_disp
```

```python
##-- Считаем дисперсию (3 компоненты объясняют 89.9% дисперсии, а 4 - 92.77%)
count = 0
sum = 0
for i in range(len(sort_disp)):
    count += 1
    sum += sort_disp[i]
    if sum > 0.9:
        break
print(count)
```

```python
trans_data = clf.transform(X)
```

```python
df1 = pd.DataFrame(trans_data)
df1
```

```python
df1[0]
```

```python
djia_data = pd.read_csv('djia_index.csv')
djia_data
```

```python
index_D = djia_data.iloc[:, 1]
```

```python
##-- Корреляция Пирсона между индексом Доу-Джонсона и обученными данными
corr = np.corrcoef(df1[0], index_D)
corr
```

```python
##-- components_ содержит информацию о том, какой вклад вносят признаки в компоненты
##-- Какая компания имеет наибольший вес в первой компоненте
prices_data.columns[np.argmax(clf.components_[0])]
```

```python
indx = -1
value = -1
for i in range(len(clf.components_[0])):
    if value < clf.components_[0][i]:
        value = clf.components_[0][i]
        indx = i
print("Company name: '%s' weight: %0.2f, index: %i" % (prices_data.columns[indx + 1], value, indx))
```
