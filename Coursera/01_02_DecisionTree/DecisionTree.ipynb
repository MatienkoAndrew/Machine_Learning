```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = pd.read_csv('titanic.csv')
data[:]
```

```python
# Оставляем пару признаков
data[['Pclass', 'Fare', 'Age', 'Sex']]
```

```python
# Избавляемся от строк с 'Nan' и 'female' заменям на '1'
X = data[['Pclass', 'Fare', 'Age', 'Sex']].dropna()
X['Sex'] = X['Sex'].map({'female': 1, 'male': 0})
X
```

```python
# Выделяем целевую переменную - выживших
Y = data.Survived
YY = data[['Pclass', 'Fare', 'Age', 'Sex', 'Survived']].dropna()
Y = YY['Survived']
```

```python
# Обучаем решающее дерево
clf = DecisionTreeClassifier(random_state=241)
clf.fit(X, Y)
importances = clf.feature_importances_
print(list(zip(X.columns, importances)))
```
