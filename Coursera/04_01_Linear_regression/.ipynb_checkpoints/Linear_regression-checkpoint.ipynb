```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import Ridge
from sklearn.feature_extraction import DictVectorizer
from scipy.sparse import hstack
```

```python
train_data = pd.read_csv('salary-train.csv')
train_data
```

```python
##-- TEST_DATA
test_data = pd.read_csv('salary-test-mini.csv')

test_data['FullDescription'] = [i.lower() for i in test_data['FullDescription']]
test_data['FullDescription'] = test_data['FullDescription'].replace('[^a-zA-Z0-9]', ' ', regex=True)

test_data['LocationNormalized'].fillna('nan', inplace=True)
test_data['ContractTime'].fillna('nan', inplace=True)
test_data
```

```python
train_data['FullDescription'] = [i.lower() for i in train_data['FullDescription']]
train_data['FullDescription'] = train_data['FullDescription'].replace('[^a-zA-Z0-9]', ' ', regex=True)
train_data['LocationNormalized'].fillna('nan', inplace=True)
train_data['ContractTime'].fillna('nan', inplace=True)
train_data
```

```python
##--  Извлечение TF-IDF-признаков
vectorizer = TfidfVectorizer(min_df=5)
X_train_vec = vectorizer.fit_transform(train_data['FullDescription'])
X_test_vec = vectorizer.transform(test_data['FullDescription'])
```

```python
enc = DictVectorizer()
X_train_categ = enc.fit_transform(train_data[['LocationNormalized', 'ContractTime']].to_dict('records'))
X_test_categ = enc.transform(test_data[['LocationNormalized', 'ContractTime']].to_dict('records'))
```

```python
X_for_train = hstack([X_train_vec, X_train_categ])
X_for_test = hstack([X_test_vec, X_test_categ])
```

```python
##-- Гребневая регрессия
clf = Ridge(alpha=1, random_state=241)
clf.fit(X_for_train, train_data['SalaryNormalized'])
```

```python
##-- Прогноз оклада
clf.predict(X_for_test)
```
